```python
import boto3
import whisper
import tempfile
import os
import subprocess
import pdfplumber
from TTS.api import TTS
from pydub import AudioSegment

MODEL = whisper.load_model("base")
TTS_MODEL = TTS("tts_models/multilingual/multi-dataset/xtts_v2")


# =========================
# PDF
# =========================
def extraer_texto_pdf(ruta_pdf):
    texto = ""
    with pdfplumber.open(ruta_pdf) as pdf:
        for p in pdf.pages:
            t = p.extract_text()
            if t:
                texto += t + "\n"
    return texto


# =========================
# DEMUCS (igual al bueno)
# =========================
def separar_audio(archivo, bucket):

    s3 = boto3.client("s3")
    nombre = os.path.splitext(os.path.basename(archivo))[0]

    subprocess.run([
        "demucs",
        "--two-stems=vocals",
        archivo
    ], check=True)

    output_folder = os.path.join("separated", "htdemucs", nombre)

    for root, _, files in os.walk(output_folder):
        for f in files:

            if f.endswith(".wav"):

                wav = os.path.join(root, f)
                mp3 = wav.replace(".wav", ".mp3")

                subprocess.run([
                    "ffmpeg",
                    "-y",
                    "-i", wav,
                    "-b:a", "192k",
                    mp3
                ], check=True)

                key = f"results/{nombre}/{f.replace('.wav','.mp3')}"
                s3.upload_file(mp3, bucket, key)


# =========================
# WHISPER (TIMING REAL)
# =========================
def transcribir_vocals(path):
    return MODEL.transcribe(path, language="en")


# =========================
# LLM (igual)
# =========================
def llamar_qwen(prompt):
    client = boto3.client("bedrock-runtime", region_name="us-east-1")

    response = client.converse(
        modelId="google.gemma-3-12b-it",
        messages=[{"role": "user", "content": [{"text": prompt}]}],
        inferenceConfig={
            "maxTokens": 1000,
            "temperature": 0.7,
            "topP": 0.9
        }
    )

    return response['output']['message']['content'][0]['text']


# =========================
# TTS (SIN CAMBIOS IMPORTANTES)
# =========================
def generar_tts(letra, vocal_ref):

    lines = [l.strip() for l in letra.split("\n") if l.strip()]
    output = []

    for i, line in enumerate(lines):

        out = f"line_{i}.wav"

        TTS_MODEL.tts_to_file(
            text=line,
            speaker_wav=vocal_ref,
            language="en",
            file_path=out
        )

        output.append(out)

    return output


# =========================
# MEZCLA (LA CLAVE DEL SONIDO BUENO)
# =========================
def mezclar(instrumental_path, segments, voices, bucket, out_key):

    s3 = boto3.client("s3")

    with tempfile.NamedTemporaryFile(suffix=".mp3", delete=False) as tmp:
        local = tmp.name

    s3.download_file(bucket, instrumental_path, local)

    instrumental = AudioSegment.from_mp3(local)
    instrumental = instrumental - 8

    final = instrumental

    max_len = min(len(segments), len(voices))

    for i in range(max_len):

        try:
            voice = AudioSegment.from_wav(voices[i])
            voice = voice + 5

            # 🔥 ESTE ES EL TIMING BUENO (WHISPER)
            start_ms = int(segments[i]["start"] * 1000)

            print(f"Overlay {i} en {start_ms}ms")

            final = final.overlay(voice, position=start_ms)

        except Exception as e:
            print(f"Error {i}: {e}")

    output = "final_song.mp3"
    final.export(output, format="mp3")

    s3.upload_file(output, bucket, out_key)

    return f"s3://{bucket}/{out_key}"


# =========================
# PIPELINE COMPLETO (VERSIÓN BUENA)
# =========================
def pipeline(pdf_bucket, pdf_key, song_path, music_bucket):

    # 1 PDF
    texto_pdf = extraer_texto_pdf(pdf_key)

    # 2 DEMUCS
    separar_audio(song_path, music_bucket)

    nombre = os.path.splitext(os.path.basename(song_path))[0]
    vocals_key = f"results/{nombre}/vocals.mp3"
    instrumental_key = f"results/{nombre}/no_vocals.mp3"

    # 3 WHISPER
    obj = boto3.client("s3").get_object(Bucket=music_bucket, Key=vocals_key)
    audio = obj["Body"].read()

    with tempfile.NamedTemporaryFile(suffix=".mp3", delete=False) as f:
        f.write(audio)
        vocal_local = f.name

    result = transcribir_vocals(vocal_local)

    segments = result["segments"]
    letra_original = result["text"]

    # 4 LLM
    prompt = f"""
Convierte este contenido en letra educativa:

PDF:
{texto_pdf}

Canción:
{letra_original}

Reglas:
- no explicaciones
- solo líneas educativas
"""

    nueva_letra = llamar_qwen(prompt)

    # 5 TTS
    voices = generar_tts(nueva_letra, vocal_local)

    # 6 MIX FINAL
    final = mezclar(
        instrumental_key,
        segments,
        voices,
        music_bucket,
        f"results/{nombre}/final_song.mp3"
    )

    return final

resultado = pipeline(
    pdf_bucket="music-project-ia",
    pdf_key="transformers_teoria.pdf",
    song_path="example3.mp3",
    music_bucket="music-project-ia"
)

print("\n===== RESULTADO FINAL =====\n")
print(resultado)
```
